Filter Data by Case

```python
france_data = df.loc[df['location'] == 'France']
```

```python
france_data['date'] = pd.to_datetime(france_data['date'])
france_data.dtypes
```

```python
france_data.set_index('date', inplace=True)
france_data = france_data.loc['2023-03']
pd.set_option('display.max_columns', 500)
france_data
```

We will calculate the probability of death (regardless of age, etc.) from the total cases in March

```python
deaths = france_data['total_deaths'].sum()
cases = france_data['total_cases'].sum()
# Case Fatality Rate
CFR = deaths / cases
```

We will try to use the population column combined with the official population data in France which is still 25 years old in march 2023

https://www.statista.com/statistics/464032/distribution-population-age-group-france/

```python
population = france_data.iloc[0]['population']
proportion_age_25 = 0.0547 #5.47 % https://www.statista.com/statistics/464032/distribution-population-age-group-france/
population_age_25 = population * proportion_age_25
```

Because the CFR that we take from total cases and total deaths is the CFR of the entire age, we will only take a partial CFR according to the total population that is still 25 years old

then we will divide the CFR by the total population of French citizens who are still 25 years old

```python
deaths_age_25 = (CFR * proportion_age_25) * population_age_25
prob_death = deaths_age_25 / population_age_25
```

```python
print("Probability of death of a person aged 25 years, citizen of France and in March 2023 : {:.10f}".format(prob_death))
```

```python
import seaborn as sns
france_data['cfr'] = france_data['total_cases'] / france_data['total_deaths']
correlation = france_data[['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'people_vaccinated',	'people_fully_vaccinated', 'cfr']].corr()
heatmap_vis = sns.heatmap(correlation, annot=True)
```

As we can see from the heatmap above, the probability of death due to covid has a strong relationship with the vaccination program. In addition, the CFR (Case Fatality Rate) column or what percentage of people who die from the occurrence of a covid case has a correlation with the Vaccination column.
